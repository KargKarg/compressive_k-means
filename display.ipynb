```python
%load_ext autoreload
%autoreload 2

from clustering import *
from sklearn.datasets import make_blobs
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 1000
D = 2
m = 500
K = 7
SEED = 12
np.random.seed(SEED)
```

```python
X, Y = make_blobs(n_samples=N, n_features=D, centers=K, random_state=SEED)
```

```python
model = CKM(K=K, m=500, lower=np.min(X), upper=np.max(X), iters=10, history=History.ON)
```

```python
model.fit(X)
```

```python
plt.scatter(x=X[:, 0], y=X[:, 1], c="black", s=0.5)
plt.scatter(x=model.centroids[:, 0], y=model.centroids[:, 1], c="red", s=20)
plt.show()
plt.close()
```
