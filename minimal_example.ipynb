```python
%load_ext autoreload
%autoreload 2

from sklearn.datasets import make_blobs
from clustering import CKM, InitCentroids, utils
import numpy as np
```

```python
N: int = 1000
D: int = 2
K: int = 5
```

```python
data: tuple[np.ndarray[float], np.ndarray[int]] = make_blobs(n_samples=N, n_features=D, centers=K, random_state=12)

X: np.ndarray[float] = data[0]
Y: np.ndarray[int] = data[1]
```

```python
model: CKM = CKM(K=K, m=N//2, lower=np.min(X), upper=np.max(X), iters=3, init_centroids=InitCentroids.SAMPLE, seed=12)
```

```python
model.fit(X)
```

```python
predicted = utils.match_labels(Y, model.predict(X))
print(predicted[:10])
print(Y[:10])
```
